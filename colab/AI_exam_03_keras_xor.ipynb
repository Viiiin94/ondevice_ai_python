```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
traning_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
target_data = np.array([[0], [1], [1], [0]])
```

```python
model = Sequential()
model.add(Dense(32 ,input_dim = 2, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss='mse', optimizer='adam', metrics=['binary_accuracy'])
model.summary()
```

2차원으론 도저히 XOR 함수를 만들 수가 없음

32차원이라는 차원으로 확장해서 XOR함수를 만듦

AND, OR는 2차원의 선으로 구분해서 만들 수가 있음

```python
fit_hist = model.fit(traning_data, target_data, epochs=500, verbose=1)
```

```python
print(model.predict(np.array([[1, 0]])))
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```
